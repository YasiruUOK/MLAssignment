```python
# https://www.kaggle.com/ronitf/predicting-heart-disease

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import itertools

# Check all the data 
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
heart_data = pd.read_csv("heart.csv")
```

```python
heart_data.shape
```

```python
heart_data.info()
```

```python
heart_data.head()
```

```python
# Check is there any duplicate rows
heart_data.duplicated(subset=None, keep='first')
# As we can see there is no any duplicate rows.
```

```python
X = heart_data.drop("target", axis=1)
X.head()
```

```python
Y = heart_data["target"]
Y.head()
```

```python
# Changing gender column
# 1-male 0-female
# we dont give chance for model to mis understand male great than female
from sklearn.preprocessing import OneHotEncoder 
onehotencoder = OneHotEncoder(categorical_features = [0]) 
data = onehotencoder.fit_transform(X[['sex']]).toarray()
data
```

```python
dataframe=pd.DataFrame(data, columns=['female', 'male'])
```

```python
X = X.join(dataframe)
X = X.drop('sex', axis = 1)
X.head()
```

```python
# Visualization of data
columns=heart_data.columns[:14]
plt.subplots(figsize=(18,15))
length=len(columns)
for i,j in itertools.zip_longest(columns,range(length)):
    plt.subplot((length/2),3,j+1)
    plt.subplots_adjust(wspace=0.2,hspace=0.5)
    heart_data[i].hist(bins=20,edgecolor='black')
    plt.title(i)
plt.show()
```

```python
X.head()
```

```python
# three fold split for better performance
from sklearn.model_selection import train_test_split
x_trainval, x_test, y_trainval, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
x_train, x_val, y_train, y_val = train_test_split(x_trainval, y_trainval, test_size = 0.2, random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = pd.DataFrame(scaler.fit_transform(x_train), columns = x_train.columns)
x_train_scaled.head()
```

```python
heart_data_train = x_train_scaled.join(y_train)
```

```python
# Lets check with the heat map also 
corrmat = heart_data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(15,15))
#plot heat map
g=sns.heatmap(heart_data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier

k_range = list(range(1,31))
weight_options = ["uniform", "distance"]

param_grid = dict(n_neighbors = k_range, weights = weight_options)
knn = KNeighborsClassifier()

from sklearn.model_selection import GridSearchCV
grid_searchknn = GridSearchCV(knn, param_grid, cv = 10, scoring = 'accuracy')
grid_searchknn.fit(x_train_scaled, y_train)
print("best gama : " +str(grid_searchknn.best_params_))
print("best accuracy : " +str(grid_searchknn.best_score_))
```

```python
knn = KNeighborsClassifier(n_neighbors = 25, weights= "uniform")
knn.fit(x_train_scaled, y_train)
```

```python
x_val_scaled = pd.DataFrame(scaler.fit_transform(x_val), columns = x_val.columns)
y_pred = knn.predict(x_val_scaled)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_val, y_pred)
```

```python
print("accuracy : {:.2f}".format(knn.score(x_val_scaled, y_val)))
```

PCA

```python
x_train_scaled.info()
```

```python
from sklearn.decomposition import PCA
# Visualize
pca = PCA()
pca.fit_transform(x_train_scaled)
pca_variance = pca.explained_variance_

plt.figure(figsize=(8, 6))
plt.bar(range(14), pca_variance, alpha=0.5, align='center', label='individual variance')
plt.legend()
plt.ylabel('Variance ratio')
plt.xlabel('Principal components')
plt.show()
```

```python
heart_data_train.info()
```

```python
pca2 = PCA(n_components=9)
pca2.fit(x_train_scaled)
x_train_pca = pca2.transform(x_train_scaled)

```

```python
from sklearn.neighbors import KNeighborsClassifier

k_range = list(range(1,31))
weight_options = ["uniform", "distance"]

param_grid = dict(n_neighbors = k_range, weights = weight_options)
knn = KNeighborsClassifier()

from sklearn.model_selection import GridSearchCV
grid_searchknn = GridSearchCV(knn, param_grid, cv = 10, scoring = 'accuracy')
grid_searchknn.fit(x_train_pca, y_train)
print("best gama : " +str(grid_searchknn.best_params_))
print("best accuracy : " +str(grid_searchknn.best_score_))
```

```python
knn = KNeighborsClassifier(n_neighbors = 14, weights= "uniform")
knn.fit(x_train_pca, y_train)
```

```python
x_val_scaled = pd.DataFrame(scaler.fit_transform(x_val), columns = x_val.columns)

pca2 = PCA(n_components=9)
pca2.fit(x_val_scaled)
x_val_pca = pca2.transform(x_val_scaled)

y_pred = knn.predict(x_val_pca)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_val, y_pred)
```

```python
print("accuracy : {:.2f}".format(knn.score(x_val_pca, y_val)))
```
